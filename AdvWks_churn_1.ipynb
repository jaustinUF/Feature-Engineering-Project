```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import DecisionTreeClassifier as DTC, plot_tree, export_text
from sklearn.metrics import classification_report, f1_score, roc_auc_score, precision_recall_curve, confusion_matrix
```

```python
df = pd.read_csv('/content/drive/MyDrive/Raffaele_Miele_mentorship/AdvWks_churn_project/Project5_with_features.csv') # get dataset
# df = pd.read_csv('/content/drive/MyDrive/Raffaele_Miele_mentorship/AdvWks_churn_project/Six_months_after.csv') # no reorders in tht six-month period
# y = df['Ordered_After_2012']                    # Define target: 1 =  ordered
y = 1 - df['Ordered_After_2012']                  # 'not' data column to define target as 1 = did not order
X = df.drop(columns=['CustomerKey', 'BirthDate', 'DateFirstPurchase', 'Ordered_After_2012'])  # Drop unnecessary columns
print(f'df shape: {df.shape}   y shape: {y.shape}')
print(f'y value counts\n{y.value_counts()}')
```

```python
# one-hot encode
categorical_cols = ['MaritalStatus', 'Gender', 'EnglishEducation', 'CommuteDistance']
X_encoded = pd.get_dummies(X, columns=categorical_cols)
```

```python
# Split the dataset into train/test sets
X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)
```

```python
print(f'y_test value counts\n{y_test.value_counts()}')
```

```python
# create and train model
model = DTC(random_state=42, max_depth=10, min_samples_split=10, min_samples_leaf=5)
model.fit(X_train, y_train)                       # train model
y_pred = model.predict(X_test)
```

```python
print(f"\nTest accuracy: {model.score(X_test, y_test) * 100:.1f}%")
print(f'\nBaseline results:\n{classification_report(y_test, y_pred)}')
```

```python
threshold = 0.5                               # (for CM heading
cm = confusion_matrix(y_test, y_pred)           # Compute the confusion matrix
cm_df = pd.DataFrame(cm, index=['no', 'yes'], columns=['no', 'yes']) # create dataframe
# Print the confusion matrix with labels
print(f"Confusion Matrix (threshold = {threshold}:\nACTUAL  PREDICTED")
print(cm_df)
```
